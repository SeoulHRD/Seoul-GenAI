```python
# 파일 경로 (업로드한 파일명 반영)
file_path = "2025년 감염병환자 발생 현황(서울시, 2025.1.1~9.17.) 수정 2.xlsx"

# 데이터 불러오기
df = pd.read_excel(file_path, sheet_name="sheet")

# 자치구별 감염병 합계 계산
df["합계"] = df.drop(columns=["자치구명"]).sum(axis=1)

# 결과 정리
result = df[["자치구명", "합계"]]

# 막대그래프 시각화
plt.figure(figsize=(14, 8))
plt.bar(result["자치구명"], result["합계"])

# 그래프 제목 및 레이블
plt.title("서울시 25개 자치구별 감염병 환자 발생 합계 (2025.1.1~9.17.)", fontsize=16)
plt.xlabel("자치구", fontsize=12)
plt.ylabel("감염병 환자 발생 합계", fontsize=12)

# x축 글자 겹침 방지
plt.xticks(rotation=45)

# 그리드 추가
plt.grid(axis="y", linestyle="--", alpha=0.7)

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
### 2. 수치형 데이터 분석 : 1개의 열
- 표 = 빈도표 : 구간의 빈도, 백분율
- 데이터 시각화 : 히스토그램, 상자그림, 바이올린
- 기술통계량 = 요약통계량
```

```python
# 파일 경로 (업로드한 파일명 반영)
file_path = "2025년 감염병환자 발생 현황(서울시, 2025.1.1~9.17.) 수정 2.xlsx"

# 데이터 불러오기
df = pd.read_excel(file_path, sheet_name="sheet")

# A형간염 자치구별 발생 현황
a_hepatitis = df[["자치구명", "A형간염"]]

# 평균 계산
mean_value = a_hepatitis["A형간염"].mean()

# 출력
print("자치구별 A형간염 발생 환자 수:")
print(a_hepatitis)
print("\n전체 평균 A형간염 발생 환자 수:", mean_value)
```

```python
# 파일 경로 (업로드한 파일명 반영)
file_path = "2025년 감염병환자 발생 현황(서울시, 2025.1.1~9.17.) 수정 2.xlsx"

# 데이터 불러오기
df = pd.read_excel(file_path, sheet_name="sheet")

# A형간염 자치구별 발생 현황
a_hepatitis = df[["자치구명", "A형간염"]]

# 평균 계산
mean_value = a_hepatitis["A형간염"].mean()

# 막대그래프 시각화
plt.figure(figsize=(14, 8))
plt.bar(a_hepatitis["자치구명"], a_hepatitis["A형간염"], color="skyblue", label="자치구별 환자 수")

# 평균선 추가
plt.axhline(mean_value, color="red", linestyle="--", linewidth=2, label=f"평균 ({mean_value:.1f})")

# 그래프 제목 및 레이블
plt.title("서울시 자치구별 A형간염 발생 환자 수 (2025.1.1~9.17.)", fontsize=16)
plt.xlabel("자치구", fontsize=12)
plt.ylabel("A형간염 발생 환자 수", fontsize=12)

# x축 글자 겹침 방지
plt.xticks(rotation=45)

# 범례 추가
plt.legend()

# 그리드 추가
plt.grid(axis="y", linestyle="--", alpha=0.7)

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
# 파일 경로 (업로드한 파일명 반영)
file_path = "2025년 감염병환자 발생 현황(서울시, 2025.1.1~9.17.) 수정 2.xlsx"

# 데이터 불러오기
df = pd.read_excel(file_path, sheet_name="sheet")

# A형간염 발생 환자수 데이터 추출
a_hepatitis = df["A형간염"]

# 시각화: 히스토그램, 상자그림, 바이올린플롯
plt.figure(figsize=(18, 5))

# 1. 히스토그램
plt.subplot(1, 3, 1)
plt.hist(a_hepatitis, bins=10, color="skyblue", edgecolor="black")
plt.title("A형간염 발생 환자수 분포 (히스토그램)")
plt.xlabel("환자수")
plt.ylabel("빈도")

# 2. 상자그림(Boxplot)
plt.subplot(1, 3, 2)
sns.boxplot(y=a_hepatitis, color="lightgreen")
plt.title("A형간염 발생 환자수 (상자그림)")
plt.ylabel("환자수")

# 3. 바이올린플롯(Violin plot)
plt.subplot(1, 3, 3)
sns.violinplot(y=a_hepatitis, color="lightcoral")
plt.title("A형간염 발생 환자수 (바이올린플롯)")
plt.ylabel("환자수")

# 레이아웃 정리
plt.tight_layout()
plt.show()
```

```python
# 파일 경로 (업로드한 파일명 반영)
file_path = "2025년 감염병환자 발생 현황(서울시, 2025.1.1~9.17.) 수정 2.xlsx"

# 데이터 불러오기
df = pd.read_excel(file_path, sheet_name="sheet")

# 전체 감염병 합계 계산
df["합계"] = df.drop(columns=["자치구명"]).sum(axis=1)

# 산점도 시각화
plt.figure(figsize=(10, 7))
plt.scatter(df["합계"], df["A형간염"], color="blue", alpha=0.7, s=80)

# 각 점에 자치구명 표시
for i, txt in enumerate(df["자치구명"]):
    plt.text(df["합계"][i] + 5, df["A형간염"][i] + 0.5, txt, fontsize=8)

# 그래프 제목 및 레이블
plt.title("서울시 자치구별 전체 감염병 환자수 vs A형간염 환자수 (산점도)", fontsize=14)
plt.xlabel("전체 감염병 환자수 합계", fontsize=12)
plt.ylabel("A형간염 환자수", fontsize=12)

# 격자 추가
plt.grid(True, linestyle="--", alpha=0.6)

# 그래프 출력
plt.tight_layout()
plt.show()
```
