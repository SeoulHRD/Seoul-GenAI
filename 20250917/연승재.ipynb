```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 엑셀 파일 불러오기
file_path = r"D:\SEOUL\K리그2024.xlsx"
df = pd.read_excel(file_path)

# X, Y 값 지정
x = df["평균관중수"]
y = df["경기당 평균득점"]

# 산점도 그리기
plt.figure(figsize=(10,6))
plt.scatter(x, y, color="blue", alpha=0.7, label="팀별 데이터")

# 각 점에 팀명 라벨 표시
for i, txt in enumerate(df["팀"]):
    plt.annotate(txt, (x[i], y[i]), fontsize=9, xytext=(5,3), textcoords="offset points")

# 회귀선(추세선) 계산
coeffs = np.polyfit(x, y, 1)   # 1차 회귀 (직선)
trendline = np.poly1d(coeffs)

# 회귀선 그리기
plt.plot(x, trendline(x), color="red", linewidth=2, label="추세선")

# 그래프 제목과 축 라벨
plt.title("K리그 2024 평균관중수 vs 경기당 평균득점", fontsize=14)
plt.xlabel("평균관중수", fontsize=12)
plt.ylabel("경기당 평균득점", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.show()
```

```python
# 엑셀 파일 불러오기
file_path = r"D:\SEOUL\K리그2024.xlsx"
df = pd.read_excel(file_path)

# X, Y 값 지정
x = df["경기당 평균득점"]
y = df["평균관중수"]

# 상관계수 계산
corr = np.corrcoef(x, y)[0, 1]

# 산점도 그리기
plt.figure(figsize=(10,6))
plt.scatter(x, y, color="blue", alpha=0.7, label="팀별 데이터")

# 각 점에 팀명 라벨 표시
for i, txt in enumerate(df["팀"]):
    plt.annotate(txt, (x[i], y[i]), fontsize=9, xytext=(5,3), textcoords="offset points")

# 1차 회귀선
coeffs1 = np.polyfit(x, y, 1)
trendline1 = np.poly1d(coeffs1)

# 2차 회귀선
coeffs2 = np.polyfit(x, y, 2)
trendline2 = np.poly1d(coeffs2)

# X 범위 설정
x_range = np.linspace(x.min(), x.max(), 200)

# 회귀선 그리기
plt.plot(x_range, trendline1(x_range), color="red", linewidth=2, label="1차 선형 추세선")
plt.plot(x_range, trendline2(x_range), color="green", linewidth=2, linestyle="--", label="2차 곡선 추세선")

# 그래프 제목과 축 라벨
plt.title("K리그 2024 경기당 평균득점 vs 평균관중수", fontsize=14)
plt.xlabel("경기당 평균득점9(골)", fontsize=12)
plt.ylabel("평균관중수(명)", fontsize=12, rotation=0, labelpad=50)  # 세로 (기본값)

# 상관계수 그래프 위에 표시
plt.text(x.min() + (x.max()-x.min())*0.05,
         y.max()*0.95,
         f"상관계수 r = {corr:.3f}",
         fontsize=12,
         bbox=dict(facecolor="white", alpha=0.6))

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.show()
```

# 추세선은 “방향성”을 보여주고, 
# 상관계수 r은 “얼마나 강한 관계가 있는지”를 수치로 보여줌

- r ≈ 0 → 관계 거의 없음
- r > 0 → 관중 수가 많을수록 득점 증가 (양의 상관관계)
- r < 0 → 관중 수가 많을수록 득점 감소 (음의 상관관계)
- |r| 값이 클수록 관계가 강함
